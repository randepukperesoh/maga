```python
class CreateDataset:
    def __init__(self, data, defect_on):
        self.class_no_defect = data
        self.Y = defect_on

    def engine_constr(self, noise_scale=0.01, num_variations=5):
        results = []

        for _ in range(num_variations):
            model = FEModel3D()

            # Добавляем узлы с возможностью внесения случайного шума в координаты
            for node in self.class_no_defect["node"]:
                perturbation_x = np.random.normal(0, 0.05)  # ±5% от исходного x
                perturbation_y = np.random.normal(0, 0.05)  # ±5% от исходного y
                model.add_node("N" + str(node["id"]), node["x"] + perturbation_x, node["y"] + perturbation_y, 0)

            # Добавляем опоры
            for node in self.class_no_defect["node"]:
                if node["supports"] == "fixed":
                    model.def_support("N" + str(node["id"]), True, True, True, True, True, True)

            # Материалы и сечения
            E = 200e3
            G = 80e3
            rho = 7850
            Fy = 250
            model.add_material('steel', E, G, rho, Fy)

            A_normal = 0.05
            I_normal = A_normal * (0.1**2) / 12
            model.add_section('S1', A_normal, I_normal, I_normal, I_normal)
            model.add_material('defect_steel', E, G, rho, Fy)
            A_defect = A_normal / 2
            I_defect = A_defect * (0.1**2) / 12
            model.add_section('S_defect', A_defect, I_defect, I_defect, I_defect)

            # Добавляем элементы с возможностью внесения случайного шума в нагрузки
            for kernel in self.class_no_defect["kernels"]:
                material = 'defect_steel' if kernel.get('defect', False) else 'steel'
                section = 'S_defect' if kernel.get('defect', False) else 'S1'
                model.add_member("K" + str(kernel["id"]), "N" + str(kernel["start_id"]), "N" + str(kernel["end_id"]), material, section)

            for kernel in self.class_no_defect["kernels"]:
                if "loadY" in kernel:
                    perturbation_load = np.random.normal(0, kernel["loadY"] * 0.1)  # ±10% от исходной нагрузки
                    model.add_member_pt_load("K" + str(kernel["id"]), "Fy", kernel["loadY"] + perturbation_load, 0.25)

            # Анализ модели
            model.analyze()

            # Собираем данные о датчиках
            data = {}
            for node in model.nodes.values():
                node_id = int(node.name[1:])
                original_node = next((n for n in self.class_no_defect['node'] if n['id'] == node_id), None)
                if original_node and original_node.get('sensor', False):
                    data[f"sensor_{original_node['id']}"] = [node.DY['Combo 1']]

            # Добавляем метку класса
            data['Y'] = [self.Y]

            # Создаем DataFrame для текущей вариации
            df_result = pd.DataFrame(data)
            results.append(df_result)

        return pd.concat(results, axis=0, ignore_index=True)


if __name__ == "__main__":
    classes_data = [
        {"data": class_no_defect, "defect_on": 0},
        {"data": class_1_1, "defect_on": 1},
        {"data": class_1_2, "defect_on": 1},
        {"data": class_1_3, "defect_on": 1},
        {"data": class_1_4, "defect_on": 1},
        {"data": class_1_5, "defect_on": 1},
        {"data": class_1_6, "defect_on": 1},
        {"data": class_1_7, "defect_on": 1},
        {"data": class_1_8, "defect_on": 1},
        {"data": class_1_9, "defect_on": 1},
        {"data": class_1_10, "defect_on": 1},
        {"data": class_1_11, "defect_on": 1},
        {"data": class_1_12, "defect_on": 1},
        {"data": class_2_1, "defect_on": 2},
        {"data": class_2_2, "defect_on": 2},
        {"data": class_2_3, "defect_on": 2},
        {"data": class_2_4, "defect_on": 2},
        {"data": class_2_5, "defect_on": 2},
        {"data": class_2_6, "defect_on": 2},
        {"data": class_2_7, "defect_on": 2},
        {"data": class_2_8, "defect_on": 2},
        {"data": class_2_9, "defect_on": 2},
        {"data": class_2_10, "defect_on": 2},
        {"data": class_2_11, "defect_on": 2},
        {"data": class_2_12, "defect_on": 2},
        {"data": class_2_13, "defect_on": 2},
        {"data": class_2_14, "defect_on": 2},
        {"data": class_3_1, "defect_on": 3},
        {"data": class_3_2, "defect_on": 3},
        {"data": class_3_3, "defect_on": 3},
        {"data": class_3_4, "defect_on": 3},
        {"data": class_3_5, "defect_on": 3},
        {"data": class_3_6, "defect_on": 3},
        {"data": class_3_7, "defect_on": 3},
        {"data": class_3_8, "defect_on": 3},
        {"data": class_3_9, "defect_on": 3},
        {"data": class_3_10, "defect_on": 3},
        {"data": class_3_11, "defect_on": 3},
        {"data": class_3_12, "defect_on": 3},
        {"data": class_3_13, "defect_on": 3},
        {"data": class_3_14, "defect_on": 3},
    ]

    results = []

    # Создаем новые вариации для каждого класса
    for class_data in classes_data:
        dataset_creator = CreateDataset(class_data["data"], class_data["defect_on"])
        result = dataset_creator.engine_constr(noise_scale=0.01, num_variations=10)  # 10 вариаций для каждого класса
        results.append(result)

    # Объединяем все результаты в один датасет
    final_result = pd.concat(results, axis=0, ignore_index=True)

    # Выводим итоговый датасет
    print(final_result)

    # Сохраняем датасет в CSV-файл
    final_result.to_csv("dataSet.csv", index=False)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python
data = pd.read_csv("dataSet.csv")

X = data[['sensor_5', 'sensor_6', 'sensor_7', 'sensor_8', 'sensor_9', 'sensor_10']].values
y = data['Y'].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
y_one_hot = to_categorical(y_encoded, num_classes=4)  # 4 класса: 0, 1, 2, 3

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_one_hot, test_size=0.2, random_state=42, stratify=y_one_hot)

model = Sequential()

model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))  # Первый скрытый слой
model.add(Dense(32, activation='relu'))  # Второй скрытый слой
model.add(Dense(4, activation='softmax'))  # Выходной слой (4 класса)

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=40, batch_size=16, validation_split=0.2, verbose=1)

test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

# model.save("defect_classification_model.h5")
```
